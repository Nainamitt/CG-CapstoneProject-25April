```python
import requests
import json
import os
from dotenv import load_dotenv

load_dotenv()

key = os.getenv("DeployKey")
url = os.getenv("DeployEndpoint")

headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {key}"
}

# The model expects these EXACT keys and data types:
data = {
    "Inputs": {
        "input1": [
            {
                "Timestamp": "2026-04-25T12:00:00Z", # Must be a string/datetime
                "MachineID": "MACH-01",              # Must be a string
                "Temperature": 85.5,                 # Must be a float/number
                "Pressure": 101.2,                  # Must be a float/number
                "Vibration": 0.02,                   # Must be a float/number
                "Humidity": 42.0                     # Must be a float/number
            }
        ]
    },
    "GlobalParameters": {}
}

try:
    # Sending the POST request
    response = requests.post(url, headers=headers, json=data)
    
    # This will catch 400, 401, or 500 errors
    response.raise_for_status() 
    
    print("Prediction Result:")
    print(json.dumps(response.json(), indent=4))

except requests.exceptions.HTTPError:
    print(f"Error 400! The model still sees a mismatch.")
    print(f"Server Response: {response.text}")
except Exception as e:
    print(f"Connection Error: {e}")
```
